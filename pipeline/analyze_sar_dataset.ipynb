## See value count for the avalible data

```python
#Load features and labels dataframe
fl_df_path = '/data/exjobb/sarssw/sar_dataset_features_labels_27_april/sar_dataset_split.pickle'
with open(fl_df_path, 'rb') as f:
    fl_df = pickle.load(f)

fl_df['any_nan'] = fl_df.isna().apply(any, axis=1)
    
display(fl_df)
display(fl_df.columns)
```

```python
print('Number of unique files grouped by non-nan values, homogenuity, swath, polarisations and split')
display(fl_df.groupby(['any_nan', 'hom_test', 'swath', 'polarisations', 'split'])['file_name'].nunique())
```

```python
display(f"proportion that contain nan values (only in swh_value from model) {fl_df['any_nan'].mean()}")
display(f"proportion of files that is passes the homogenuity filter {fl_df['hom_test'].mean()}")
```

## See number of files filtered based on groupby on file_name and polarisations

```python
#Load features and labels dataframe
fl_df_path = '/data/exjobb/sarssw/sar_dataset_features_labels_27_april/sar_dataset_split.pickle'
with open(fl_df_path, 'rb') as f:
    fl_df = pickle.load(f)

fl_df['any_nan_row_wise'] = fl_df.isna().apply(any, axis=1)
grouped = fl_df.groupby(['file_name', 'polarisations'])

fl_df['any_nan_group_wise'] = grouped['any_nan_row_wise'].transform("any")
fl_df['hom_test_group_wise'] = grouped['hom_test'].transform("all")
    
display(fl_df)
display(fl_df.columns)
```

```python
print("""
Here the data is grouped per file (using columns file_name and polarisations)
Number of unique files grouped by any_nan (group wise) values, homogenuity (group wise), swath, polarisations and split
""")
display(fl_df.groupby(['any_nan_group_wise', 'hom_test_group_wise', 'swath', 'polarisations', 'split'])['file_name'].nunique())
```
