## See value count for the avalible data

```python
#Load features and labels dataframe
fl_df_path = '/data/exjobb/sarssw/sar_dataset_features_labels_27_april/sar_dataset_split.pickle'
with open(fl_df_path, 'rb') as f:
    fl_df = pickle.load(f)

fl_df['any_nan'] = fl_df.isna().apply(any, axis=1)
    
display(fl_df)
display(fl_df.columns)
```

```python
# Value counts for Homogenuity, non-nan, split, mode and polarisering
fl_df_unique = fl_df.drop_duplicates(['file_name', 'polarisations'])

display(fl_df_unique.value_counts(subset=['any_nan', 'hom_test', 'split', 'swath', 'polarisations']))
```

```python
display(f"proportion that contain nan values (only in swh_value from model) {fl_df['any_nan'].mean()}")
display(f"proportion of files that is passes the homogenuity filter {fl_df['hom_test'].mean()}")
```
