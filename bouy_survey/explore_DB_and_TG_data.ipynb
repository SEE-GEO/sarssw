This notebook analyses the DB and TG datasets

Conclusion: 
DB only contains the data variables ['DC_REFERENCE', 'DEPH', 'TEMP', 'ATMS', 'ATPT']
TG only contains the data variables ['DC_REFERENCE', 'DEPH', 'SLEV']

```python
import numpy as np
import pandas as pd
import xarray as xr
import os
from tqdm import tqdm
```

### Explore DB data

```python
data_dir_DB = '/data/exjobb/sarssw/bouy/INSITU_GLO_PHYBGCWAV_DISCRETE_MYNRT_013_030/DB'
files_DB = os.listdir(data_dir_DB)
data_file_DB = files_DB[0]
f_DB = xr.open_dataset(os.path.join(data_dir_DB, data_file_DB))
f_DB
```

```python
variable_filter_DB = ['DC_REFERENCE', 'DEPH', 'TEMP', 'ATMS', 'ATPT']
interesting_files_DB = {}
for file in tqdm(files_DB):
    f = xr.open_dataset(os.path.join(data_dir_DB, data_file_DB))
    interesting_vars = [l for l in f.data_vars if not 'QC' in l and l not in variable_filter_DB]
    if len(interesting_vars) != 0:
        print(file, interesting_vars)
        interesting_files_DB[file] = interesting_vars
```
